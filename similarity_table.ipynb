```python
import pandas as pd

recipe_df = pd.read_csv("csv/full_dataset.csv", header=0, index_col=0)
print(recipe_df.shape)

NER = recipe_df[['title', 'NER']]
```

```python
def str_to_list(s): return s[2:-2].split('", "')

def select_matching(row, n=6):
    title, ings = row
    matching = []
    for index2, selected_row2 in NER.iterrows():
        title2, ings2 = selected_row2
        common = n_matching(str_to_list(ings), str_to_list(ings2))
        if len(common) >= n:
            # print(selected_row2['NER'])
            # print(f'{title}: {selected_row2['title']} {common}')
            matching.append(index2)
    return matching

def n_matching(l1, l2):
    common = []
    for ing in l1:
        if ing in l2:
            common.append(ing)
    return common
```

```python
def select_largest_non_matching(row):
    title, ings = row
    longest = []
    for index2, selected_row2 in NER.iterrows():
        title2, ings2 = selected_row2
        common = n_matching(str_to_list(ings), str_to_list(ings2))
        if len(common) == 0 and len(str_to_list(ings2)) == 10:
            longest = str_to_list(ings2)
    return longest
```

```python
print(select_largest_non_matching(NER.iloc[6,:]))
```

```python
rows = [5,6,13,19]
NER.iloc[rows,:]
```

```python
match_col = []
for row in rows:
    matching = select_matching(NER.iloc[row,:], n=8)
    match_col.append(matching)
match_col
```

```python
similarity = NER.iloc[rows,:]
similarity.insert(len(NER.columns), 'similar_recipes',  match_col)
similarity
```
