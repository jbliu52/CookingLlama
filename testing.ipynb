# Evaluation

```python
import numpy as np
import math
import matplotlib.pyplot as plt

def get_accuracies(dir):
    accuracies = {}
    for model in openai_list: accuracies[model] = 0
    for model in together_list: accuracies[model] = 0

    metadata = pd.read_csv(f'{dir}/metadata.csv')
    outputs = pd.read_csv(f'{dir}/outputs.csv')
    for col in outputs.columns:
        raw = [text for text in outputs.loc[:, col]]
        correct = [(r.rfind(metadata['correct'][i]) != -1 and
                    r.rfind('Recipe') == r.rfind(metadata['correct'][i]))
                   for i, r in enumerate(raw)]

        mean = len(outputs.loc[:, col][correct]) / len(raw)
        sd = math.sqrt((mean * (1-mean)) / len(raw))
        accuracies[col] = mean, sd

    return accuracies

def plot_comp(dir1, dir2, labels):
    acc1 = get_accuracies(dir1)
    acc2 = get_accuracies(dir2)

    means = {}
    sds = {}
    for label in labels:
        means[label] = []
        sds[label] = []

    for model in acc1:
        mean1, sd1 = acc1[model]
        mean2, sd2 = acc2[model]
        means[labels[0]].append(mean1)
        sds[labels[0]].append(sd1)
        means[labels[1]].append(mean2)
        sds[labels[1]].append(sd2)


    x = np.arange(len(acc1))  # the label locations
    width = 0.25  # the width of the bars
    multiplier = 0

    fig, ax = plt.subplots(layout='constrained')

    for label in means:
        offset = width * multiplier
        ax.bar(x + offset, means[label], width, edgecolor='black', label=label)
        ax.errorbar(x + offset, means[label], yerr = sds[label], fmt ='o', color='black')
        multiplier += 1

    ax.set_xticks(range(len(acc1)), [model.split('/')[-1] for model in acc1.keys()], rotation=30)
    ax.set_ylim(0,1)

    plt.title(f'{labels[0]} vs. {labels[1]}')
    plt.legend()
    plt.show()

def plot_line_logx(ex:list):
    means = {}
    sds = {}
    for model in openai_list:
        means[model] = []
        sds[model] = []
    for model in together_list:
        means[model] = []
        sds[model] = []

    xticks = [int(d.split('_')[1][1:]) for d in ex]
    xticks.sort()

    for i in np.argsort([int(d.split('_')[1][1:]) for d in ex]):
        metadata = pd.read_csv(f'{ex[i]}/metadata.csv')
        outputs = pd.read_csv(f'{ex[i]}/outputs.csv')
        for col in outputs.columns:
            raw = [text for text in outputs.loc[:, col]]
            correct = [(r.rfind(metadata['correct'][j]) != -1 and
                    r.rfind('Recipe') == r.rfind(metadata['correct'][j]))
                   for j, r in enumerate(raw)]

            mean = len(outputs.loc[:, col][correct]) / len(raw)
            sd = math.sqrt((mean * (1-mean)) / len(raw))
            # accuracies[col] = mean, sd
            means[col].append(mean)
            sds[col].append(sd)

    # print(means)
    # print(sds)

    import matplotlib.pyplot as plt

    fig, ax = plt.subplots(layout='constrained', figsize=(15, 8))

    for i, model in enumerate(means):
        mean = np.array(means[model])
        sd = np.array(sds[model])
        ax.plot(xticks, mean, label=model.split('/')[-1])
        # ax.errorbar(range(len(mean)), mean, yerr = sd, fmt ='o', color='black')
        ax.fill_between(xticks, mean-sd, mean+sd ,alpha=0.3)

    ax.set_xlim(2,128)
    ax.set_ylim(0,1)
    ax.set_title('Results')
    ax.set_xscale('log')
    ax.set_xticks(xticks, [f'{x} options' for x in xticks], rotation=30)
    ax.grid(True)

    fig.legend(loc='outside right upper')
    plt.show()


def plot_line(ex:list):
    means = {}
    sds = {}
    for model in openai_list:
        means[model] = []
        sds[model] = []
    for model in together_list:
        means[model] = []
        sds[model] = []

    xticks = [float(d.split('_p')[1]) for d in ex]
    xticks.sort()

    for i, folder in enumerate(ex):
        metadata = pd.read_csv(f'{ex[i]}/metadata.csv')
        outputs = pd.read_csv(f'{ex[i]}/outputs.csv')
        for col in outputs.columns:
            raw = [t for t in outputs.loc[:, col]]
            correct = [(str(r).rfind(metadata['correct'][j]) != -1 and
                    str(r).rfind('Recipe') == str(r).rfind(metadata['correct'][j]))
                   for j, r in enumerate(raw)]

            mean = len(outputs.loc[:, col][correct]) / len(raw)
            sd = math.sqrt((mean * (1-mean)) / len(raw))
            # accuracies[col] = mean, sd
            means[col].append(mean)
            sds[col].append(sd)

    # print(means)
    # print(sds)

    import matplotlib.pyplot as plt

    fig, ax = plt.subplots(layout='constrained', figsize=(15, 8))

    for i, model in enumerate(means):
        mean = np.array(means[model])
        sd = np.array(sds[model])
        ax.plot(xticks, mean, label=model.split('/')[-1])
        # ax.errorbar(range(len(mean)), mean, yerr = sd, fmt ='o', color='black')
        ax.fill_between(xticks, mean-sd, mean+sd ,alpha=0.3)

    ax.set_xlim(0,1)
    ax.set_ylim(0,1)
    ax.set_title('Results')
    ax.set_xticks(xticks, [f'{x*100}% irrelevant options' for x in xticks], rotation=30)
    ax.grid(True)

    fig.legend(loc='outside right upper')
    plt.show()
```

# Patching

```python
dataroot = 'results/t4_n64_l4_s0_(8,16)'
model = 'Qwen/Qwen3.5-397B-A17B'

meta_df = pd.read_csv(f'{dataroot}/metadata.csv')
out_df = pd.read_csv(f'{dataroot}/outputs.csv')
full_df = pd.read_csv(f'{dataroot}/full_outputs.csv')

qwen_col = out_df.loc[:, model]

timeout_inds = [i for i, out in enumerate(qwen_col) if out == 'timeout']
for i in tqdm(timeout_inds):
    prompt = meta_df.loc[i, 'full_prompt']
    client = Together()
    try:
        response = client.chat.completions.create(
            model=model,
            messages=[
                # {"role": "user", "content": instructions},
                # {"role": "user", "content": instructions + transcript_text + mistake_text}
                {"role": "user", "content": prompt}
            ],
            reasoning={"enabled": True},
            max_tokens=30000,
            timeout=300
        )
        out_df.loc[i, model] = response.choices[0].message.content
        full_df.loc[i, model] = response
    except Exception as e: print(f'Error: {str(e)} ({model})')

out_df.to_csv(f'{dataroot}/outputs.csv', index=False)
full_df.to_csv(f'{dataroot}/full_outputs.csv', index=False)
```
