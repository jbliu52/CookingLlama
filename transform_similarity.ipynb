```python
def tr_match(r1, r2):
    try:
        t1list = r1.split('\n')
        t2list = r2.split('\n')

        def match_type(t1, t2):
            m1 = t1.split("', '")[0].split("'")[1]
            m2 = t2.split("', '")[0].split("'")[1]
            return m1 == m2

        def match_ings(t1, t2):
            l1 = t1.split("]")[0].split("[")[1].split(', ')
            l2 = t2.split("]")[0].split("[")[1].split(', ')
            l1.sort()
            l2.sort()
            return l1 == l2

        total = 0
        for t1 in t1list:
            for t2 in t2list:
                if match_type(t1, t2) and match_ings(t1, t2): total += 1

        return total
    except:
        return 0
```

```python
import pandas as pd
import numpy as np

recipe_df = pd.read_csv("csv/cake_recipes_formatted.csv", header=0, index_col=0)
print(recipe_df.shape)
```

```python
pairs = []
matching = []

for title1, row1 in recipe_df.iterrows():
    for title2, row2 in recipe_df.iterrows():
        if title1 == title2: continue
        pairs.append((title1, title2))
        matching.append(tr_match(row1[4], row2[4]))
```

```python
pairs = np.array(pairs)[-np.argsort(matching)]
matching.sort(reverse=True)
print(pairs)
print(matching)
```
